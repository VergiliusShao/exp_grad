```python
import numpy as np
import matplotlib.pyplot as plt
import os
print(os.getcwd())
```

```python
file = open("logistic_AdaGrad.csv")
reg_adagrad = np.loadtxt(file, delimiter=",")
file = open("logistic_AdaFTRL.csv")
reg_adaftrl = np.loadtxt(file, delimiter=",")
file = open("logistic_AdaHU.csv")
reg_adahu = np.loadtxt(file, delimiter=",")
file = open("logistic_AdaExpGrad.csv")
reg_adaexpgrad = np.loadtxt(file, delimiter=",")
file = open("logistic_AdaExpFTRL.csv")
reg_adaexpftrl = np.loadtxt(file, delimiter=",")
```

```python
plt.plot(range(0,10000), reg_adagrad,label='MD', marker='v',markevery=1000)
plt.plot(range(0,10000), reg_adaftrl,label='FTRL', marker='^',markevery=1000)
plt.plot(range(0,10000), reg_adahu,label='HU', marker='s',markevery=1000)
plt.plot(range(0,10000), reg_adaexpgrad,label='Exp-MD', marker='o',markevery=1000)
plt.plot(range(0,10000), reg_adaexpftrl,label='Exp-FTRL', marker='D',markevery=1000)
plt.legend()
plt.xlabel("Iteration")
plt.ylabel("Regret")
```
